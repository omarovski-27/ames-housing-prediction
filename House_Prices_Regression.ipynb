# Feature Engineering

```python
# We will create a new feature combining square footage and call it TotalSF
df["TotalSF"]=df["1stFlrSF"]+df["2ndFlrSF"]+df["TotalBsmtSF"]
# Check the correlation with sales price
print("The correlation with sales price is :", df["TotalSF"].corr(df["SalePrice"]))
```

```python
# Total bathrooms

# Create TotalBath: Full baths count as 1, Half baths count as 0.5
df["TotalBath"] = (
    df["FullBath"] + 
    df["BsmtFullBath"] + 
    (0.5 * df["HalfBath"]) + 
    (0.5 * df["BsmtHalfBath"])
)

# Check the correlation!
print("TotalBath Correlation:", df["TotalBath"].corr(df["SalePrice"]))
```

```python
#Building age
df["HouseAge"]=df["YrSold"]-df["YearBuilt"]
print("HouseAge Correlation:", df["HouseAge"].corr(df["SalePrice"]))
```

```python
# We first need to know the columns that are in text
text_cols=df.select_dtypes(include=['object']).columns
print(text_cols)
```

```python
# 1. Functional: Rank from Salvage (0) to Typical (7) - These are the ordinal columns
functional_map = {
    'Typ': 7, 'Min1': 6, 'Min2': 5, 'Mod': 4, 
    'Maj1': 3, 'Maj2': 2, 'Sev': 1, 'Sal': 0
}
df['Functional'] = df['Functional'].map(functional_map)

# 2. LotShape: Regular (3) to Irregular (0)

lot_shape_map = {'Reg': 3, 'IR1': 2, 'IR2': 1, 'IR3': 0}
df['LotShape'] = df['LotShape'].map(lot_shape_map)

# 3. PavedDrive: Yes (2) > Partial (1) > No (0)
paved_map = {'Y': 2, 'P': 1, 'N': 0}
df['PavedDrive'] = df['PavedDrive'].map(paved_map)

# 4. LandSlope: Gentle (2) > Moderate (1) > Severe (0)
slope_map = {'Gtl': 2, 'Mod': 1, 'Sev': 0}
df['LandSlope'] = df['LandSlope'].map(slope_map)
 
# Check if any new missing values appeared 
print("Missing after mapping:", df[['Functional', 'LotShape', 'PavedDrive', 'LandSlope']].isnull().sum().sum())
```

```python
# Now we one-hot encode the nominal columns
df=pd.get_dummies(df,drop_first=True)
print("New Shape: ",df.shape)
print("Text columns left:", df.select_dtypes(include=['object']).shape[1])
```

# Building the model - Linear Regression

```python
# We will use multiple models and compare, we start off with linear regression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X=df.drop(["SalePrice","Id"],axis=1)
y = df["SalePrice"]  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model=LinearRegression()
scaler=StandardScaler()
scaled_X_train= scaler.fit_transform(X_train)
scaled_X_test=scaler.transform(X_test)
model.fit(scaled_X_train,y_train)
pred=model.predict(scaled_X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
mae=mean_absolute_error(y_test,pred)
mse=mean_squared_error(y_test,pred)
rmse=mse**0.5
print(f"The Mean Absolute Error is: {mae}")
print(f"The Mean Squared Error is: {mse}")
print(f"The Root Mean Squared Error is: {rmse}")
```

#### We notice massive overfitting since linear model tried to fit every single number which made the errors explode

# Ridge Regression

```python
from sklearn.linear_model import Ridge
model=Ridge()
model.fit(scaled_X_train,y_train)
pred=model.predict(scaled_X_test)
mae=mean_absolute_error(y_test,pred)
mse=mean_squared_error(y_test,pred)
rmse=mse**0.5
print(f"The Mean Absolute Error is: {mae}")
print(f"The Mean Squared Error is: {mse}")
print(f"The Root Mean Squared Error is: {rmse}")

```

#### Big numbers, here the RMSE punishes the error by almost double the value, this indicates outliers missing with the data

```python
# Attempt to tune hyperparameter "Alpha" in Ridge Regression
alphas = [0.1, 1, 10, 100]

for a in alphas:
    # 1. Create model with specific alpha
    ridge = Ridge(alpha=a)
    
    # 2. Train
    ridge.fit(scaled_X_train, y_train)
    
    # 3. Predict & Score
    p = ridge.predict(scaled_X_test)
    score = np.sqrt(mean_squared_error(y_test, p))
    
    print(f"Alpha: {a} | RMSE: ${score}")
```

#### Huge improvement for Alpha=100

# Random Forest Regressors

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=100,random_state=42)
model.fit(scaled_X_train,y_train)
pred=model.predict(scaled_X_test)
rmse = np.sqrt(mean_squared_error(y_test, pred))
print(f"Basic Forest RMSE: ${rmse:,.2f}")

```

#### Random Forest is working better!

```python
# Lets use gridsearch to find the best hyperparameters
from sklearn.model_selection import GridSearchCV
param_grid={"n_estimators":[100,150,170,200,230],
            "max_depth":[5,7,10,15]}
grid=GridSearchCV(estimator=model,param_grid=param_grid,cv=3,scoring="neg_root_mean_squared_error",verbose=1,n_jobs=-1)
print("Training... (This might take 1-2 minutes)")
grid.fit(scaled_X_train, y_train)
print("Best Params:", grid.best_params_)
print("Best CV Score:", grid.best_score_)
```

#### A little insight here which is that this shows that we got lucky with the basic forest RMSE, and here we used 5 folds and it averaged 31k.

```python
# 1. Grab the best model automatically
final_forest = grid.best_estimator_

# 2. Predict on the Test Set (X_test)
final_pred = final_forest.predict(scaled_X_test)

# 3. Final Score
final_mae = mean_absolute_error(y_test, final_pred)
final_rmse = np.sqrt(mean_squared_error(y_test, final_pred))

print(f"Final Tuned Forest MAE:  ${final_mae:,.2f}")
print(f"Final Tuned Forest RMSE: ${final_rmse:,.2f}")
```

# Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
model=GradientBoostingRegressor(n_estimators=100, learning_rate=0.1,max_depth=3,random_state=42)
model.fit(scaled_X_train,y_train)
pred=model.predict(scaled_X_test)
gb_rmse = np.sqrt(mean_squared_error(y_test, pred))

print(f"Gradient Boosting RMSE: ${gb_rmse:,.2f}")
print(f"Random Forest Best:     $26,225 (Target to beat)")
```

```python
from sklearn.model_selection import GridSearchCV

gb = GradientBoostingRegressor(random_state=42)

param_grid = {
    "n_estimators": [50,70,100,150,200,250,300],
    "learning_rate": [0.05, 0.1, 0.2],
    "max_depth": [3, 4, 5]
}

gb_search = GridSearchCV(
    estimator=gb,
    param_grid=param_grid,
    cv=3,
    scoring="neg_root_mean_squared_error",
    verbose=1,
    n_jobs=-1
)

print("Tuning Gradient Booster...")
gb_search.fit(scaled_X_train, y_train)

print("Best Params:", gb_search.best_params_)
print("Best CV Score:", -gb_search.best_score_)
```

```python
final_gb = gb_search.best_estimator_

final_pred = final_gb.predict(scaled_X_test)

final_rmse = np.sqrt(mean_squared_error(y_test, final_pred))

print(f"Tuned GB Score (X_test): ${final_rmse:,.2f}")
print(f"Default GB Score:        $24,163.70")
```

```python
# Reload and process test data from scratch
test_df = pd.read_csv("test.csv")
test_ids = test_df['Id']
test_df = test_df.drop('Id', axis=1)

# Safety net fills
math_cols = ["TotalBsmtSF", "BsmtUnfSF", "BsmtFinSF1", "BsmtFinSF2", "BsmtFullBath", "BsmtHalfBath", "GarageCars", "GarageArea"]
for col in math_cols:
    test_df[col] = test_df[col].fillna(0)

test_df["Functional"] = test_df["Functional"].fillna("Typ")
test_df["KitchenQual"] = test_df["KitchenQual"].fillna("TA")
test_df["MSZoning"] = test_df["MSZoning"].fillna(test_df["MSZoning"].mode()[0])
test_df["SaleType"] = test_df["SaleType"].fillna(test_df["SaleType"].mode()[0])
test_df["Utilities"] = test_df["Utilities"].fillna("AllPub")
test_df["Exterior1st"] = test_df["Exterior1st"].fillna(test_df["Exterior1st"].mode()[0])
test_df["Exterior2nd"] = test_df["Exterior2nd"].fillna(test_df["Exterior2nd"].mode()[0])

# Quality mappings
quality_map = {"Ex":5, "Gd":4, "TA":3, "Fa":2, "Po":1}
quality_cols = ["ExterQual","HeatingQC","KitchenQual","FireplaceQu","GarageQual","PoolQC","ExterCond","BsmtQual","BsmtCond","GarageCond"]
for col in quality_cols:
    test_df[col] = test_df[col].map(quality_map).fillna(0)

garage_finish_map = {"Fin":3, "RFn":2, "Unf":1}
test_df["GarageFinish"] = test_df["GarageFinish"].map(garage_finish_map).fillna(0)

BsmtExposure_map = {"Gd":4, "Av":3, "Mn":2, "No":1}
test_df["BsmtExposure"] = test_df["BsmtExposure"].map(BsmtExposure_map).fillna(0)

bsmt_map = {"GLQ":6, "ALQ":5, "BLQ":4, "Rec":3, "LwQ":2, "Unf":1}
for col in ['BsmtFinType1', 'BsmtFinType2']:
    test_df[col] = test_df[col].map(bsmt_map).fillna(0)

# Nominal fills
nominal_cols = ["Alley","Fence","MiscFeature","GarageType","MasVnrType"]
for col in nominal_cols:
    test_df[col] = test_df[col].fillna("None")

test_df["LotFrontage"] = test_df["LotFrontage"].fillna(test_df.groupby("Neighborhood")["LotFrontage"].transform("median"))
test_df["MasVnrArea"] = test_df["MasVnrArea"].fillna(0)
test_df["Electrical"] = test_df["Electrical"].fillna(test_df["Electrical"].mode()[0])
test_df["GarageYrBlt"] = test_df["GarageYrBlt"].fillna(0)

# Feature engineering
test_df["TotalSF"] = test_df["1stFlrSF"] + test_df["2ndFlrSF"] + test_df["TotalBsmtSF"]
test_df["TotalBath"] = test_df["FullBath"] + test_df["BsmtFullBath"] + (0.5 * test_df["HalfBath"]) + (0.5 * test_df["BsmtHalfBath"])
test_df["HouseAge"] = test_df["YrSold"] - test_df["YearBuilt"]

# Ordinal mappings
functional_map = {'Typ':7, 'Min1':6, 'Min2':5, 'Mod':4, 'Maj1':3, 'Maj2':2, 'Sev':1, 'Sal':0}
test_df['Functional'] = test_df['Functional'].map(functional_map).fillna(7)

lot_shape_map = {'Reg':3, 'IR1':2, 'IR2':1, 'IR3':0}
test_df['LotShape'] = test_df['LotShape'].map(lot_shape_map).fillna(0)

paved_map = {'Y':2, 'P':1, 'N':0}
test_df['PavedDrive'] = test_df['PavedDrive'].map(paved_map).fillna(0)

slope_map = {'Gtl':2, 'Mod':1, 'Sev':0}
test_df['LandSlope'] = test_df['LandSlope'].map(slope_map).fillna(0)

# One-hot encode and align columns
test_df = pd.get_dummies(test_df, drop_first=True)
final_test_data = test_df.reindex(columns=X.columns, fill_value=0)

# Predict
scaled_test_data = scaler.transform(final_test_data)
pred = final_gb.predict(scaled_test_data)

submission = pd.DataFrame({
    "Id": test_ids,
    "SalePrice": pred
})

submission.to_csv("submission.csv", index=False)

print("Min price:", pred.min())
print("Max price:", pred.max())
print("Mean price:", pred.mean())
```
